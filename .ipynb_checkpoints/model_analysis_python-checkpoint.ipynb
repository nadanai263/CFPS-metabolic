```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib.ticker import AutoMinorLocator
from matplotlib.ticker import MaxNLocator
```

## 1. Plot traces

```python
def formatplot(ax,xlabel,ylabel,xlim=False,ylim=False,logx=False,logy=False,logxy=False,symlogx=False,padx=12,pady=12):

    ######### SET TITLES AND LABLES #########

    #ax.set_title('Plot title')
    if xlabel!=False:
        ax.set_xlabel(xlabel, labelpad=padx)
    if ylabel!=False:    
        ax.set_ylabel(ylabel, labelpad=pady)

    ######### SET AXES LIMITS #########

    if xlim!=False:
        ax.set_xlim(xlim)
    if ylim!=False:
        ax.set_ylim(ylim)

    ######### SET TICK VALUES #########

    #ax.set_xticks([0,0.5,1])
    #	ax.set_yticks([0,2,4,6,8])

    ######### SET LINE THICKNESSES #########


    ax.spines['bottom'].set_linewidth(1.5)
    ax.spines['top'].set_linewidth(1.5)
    ax.spines['left'].set_linewidth(1.5)
    ax.spines['right'].set_linewidth(1.5) 

    ######### SET TICKS #########

    if logx==True:

        ax.set_xscale("log")

    elif logy==True:

        ax.set_yscale("log")

    elif logxy==True:

        ax.set_xscale("log")
        ax.set_yscale("log")
    
    elif symlogx==True:

        ax.set_xscale("symlog",linthreshx=1e-4)
        ax.set_yscale("log")

    else:
        minorLocatorx=AutoMinorLocator(2) # Number of minor intervals per major interval
        minorLocatory=AutoMinorLocator(2)
        ax.xaxis.set_minor_locator(minorLocatorx)
        ax.yaxis.set_minor_locator(minorLocatory)

    ax.tick_params(which='major', width=1.5, length=8, pad=9,direction='in',top='on',right='on')
    ax.tick_params(which='minor', width=1.5, length=4, pad=9,direction='in',top='on',right='on')
```

```python
glu = pd.read_csv('./2Dscan/exp_glu.csv')
CAT = pd.read_csv('./2Dscan/exp_CAT.csv')
pyr = pd.read_csv('./2Dscan/exp_pyr.csv')
lac = pd.read_csv('./2Dscan/exp_lac.csv')

xglu = glu['x1']
yglu = glu['x2']
yerrglu = glu['x3']

xCAT = CAT['x1']
yCAT = CAT['x2']
yerrCAT = CAT['x3']

xpyr = pyr['x1']
ypyr = pyr['x2']
yerrpyr = pyr['x3']

xlac = lac['x1']
ylac = lac['x2']
yerrlac = lac['x3']

model = pd.read_csv("./2Dscan/refoutput.csv")
modeltime = pd.read_csv("./2Dscan/refoutput_t.csv")
```

```python
plt.close("all")

my_dpi=150

figure_options={'figsize':(15,3)} #figure size in inches. A4=11.7x8.3. A5=8.27,5.83
font_options={'size':'14','family':'sans-serif','sans-serif':'Arial'}
plt.rc('figure', **figure_options)
plt.rc('font', **font_options)

#current_palette=sns.color_palette("coolwarm", len(c))
#plt.rc('axes',prop_cycle=(cycler('color',current_palette)))
f, axarr=plt.subplots(1,4)
plt.subplots_adjust(left=0.25,bottom=0.2,right=0.95,top=0.95)

axarr[0].errorbar(xglu,yglu,yerr=yerrglu,fmt='ko',ms=5,elinewidth=1.5,capsize=5,mew=1.5)
axarr[1].errorbar(xCAT,yCAT*1000,yerr=yerrCAT*1000,fmt='ko',ms=5,elinewidth=1.5,capsize=5,mew=1.5)
axarr[2].errorbar(xpyr,ypyr,yerr=yerrpyr,fmt='ko',ms=5,elinewidth=1.5,capsize=5,mew=1.5)
axarr[3].errorbar(xlac,ylac,yerr=yerrlac,fmt='ko',ms=5,elinewidth=1.5,capsize=5,mew=1.5)

axarr[0].plot(modeltime.iloc[:,0],model.iloc[99,1:],'b',lw=1.5)
axarr[1].plot(modeltime.iloc[:,0],model.iloc[97,1:]*1000,'b',lw=1.5)
axarr[2].plot(modeltime.iloc[:,0],model.iloc[101,1:],'b',lw=1.5)
axarr[3].plot(modeltime.iloc[:,0],model.iloc[103,1:],'b',lw=1.5)

formatplot(axarr[0],'Time (h)', 'glucose (mM)',False,False)
formatplot(axarr[1],'Time (h)', 'CAT ($\mu$M)',False,False)
formatplot(axarr[2],'Time (h)', 'pyruvate (mM)',False,False)
formatplot(axarr[3],'Time (h)', 'lactate (mM)',False,False)

f.tight_layout()
plt.savefig('./2Dscan/plots/data.pdf',dpi=my_dpi,transparent=True)
```

## 2. Plot heatmaps

```python
upup = pd.read_csv('./2Dscan/upup.csv')
updwn = pd.read_csv('./2Dscan/updwn.csv')
dwndwn = pd.read_csv('./2Dscan/dwndwn.csv')
```

```python
idx2d = pd.read_csv('./2Dscan/idx2d.csv').values
idx = np.array([j for i in idx2d for j in i])
idx0 = idx-1 # change to 0-index
```

```python
upupsel = upup.values[idx0[:,None],idx0]
updwnsel = updwn.values[idx0[:,None],idx0]
dwndwnsel = dwndwn.values[idx0[:,None],idx0]
```

```python
plt.close("all")

my_dpi=150

figure_options={'figsize':(18,14)} #figure size in inches. A4=11.7x8.3. 
font_options={'size':'28','family':'sans-serif','sans-serif':'Arial'}
plt.rc('figure', **figure_options)
plt.rc('font', **font_options)
fig=plt.figure(); ax=fig.add_subplot(1,1,1)

xticklabels=idx
yticklabels=idx

ax=sns.heatmap(upupsel/0.01856062311401059,cbar=True,cmap='coolwarm_r',square=True, 
               yticklabels=yticklabels, xticklabels=xticklabels,
               vmax=2, vmin=0) #vmax=7
plt.yticks(rotation=0)
plt.xticks(rotation=90)
ax.tick_params(labelsize=28)
ax.set_xlabel('Enzyme 1, 10x',labelpad=9,size=40)
ax.set_ylabel('Enzyme 2, 10x',labelpad=9,size=40)


plt.savefig('./2Dscan/plots/upup.pdf',dpi=my_dpi,bbox_inches='tight')
```
